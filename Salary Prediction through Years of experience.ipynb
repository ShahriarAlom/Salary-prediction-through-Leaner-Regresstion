```python
import pandas as pd
from pandas import DataFrame
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
%matplotlib inline
```

```python
data = pd.read_csv('Salary_dataset.csv')
data.info()
```

```python
data = pd.DataFrame(data)
```

```python
df = data[['YearsExperience', 'Salary']]
X = data[['YearsExperience']]
y = data[['Salary']]
```

```python
#df
```

```python
#data.head()
data.tail()
```

```python
plt.scatter(X, y, c = '#B324FF')
plt.title('YearsExperience vs Salary')
plt.xlabel('YearsExperience')
plt.ylabel('Salary')
plt.xlim(0,11)
plt.ylim(0, 130000)
plt.show()
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, 
                                                    test_size=0.15, random_state=10)
```

```python
regr = LinearRegression()
```

```python
regr.fit(X_train,y_train)
```

```python
# m is called coeficient or slope
# c is called intercept or constant(slope starting point)
m = regr.coef_[0][0] 
c = regr.intercept_[0]
r_square = regr.score(X,y)
print('m =', m)
print('c =', c)
print('r-square =', r_square)
```

```python
predicted_value = regr.predict(X_train)
```

```python
%matplotlib inline
plt.scatter(X_train, y_train, c = '#B324FF')
plt.plot(X_train, predicted_value) #this is the code for predicted line (this is for straight line)
plt.scatter(X_train, predicted_value, c= "black") #this is for specific predicted points
plt.title('YearsExperience vs Salary')
plt.xlabel('YearsExperience')
plt.ylabel('Salary')
plt.xlim(0,11)
plt.ylim(0, 130000)
plt.style
plt.show()
```
